```python
data = ('col1,col2,col3\n'
       'a,b,c\n'
       'd,e,f')
```

```python
from io import StringIO
```

```python
pd.read_csv(StringIO(data))
```

```python
## With the help of parameters of read_csv, we can make pandas read only specific columns
pd.read_csv(StringIO(data),usecols = ['col1','col3'])
```

```python
int_data = ('col1,col2,col3\n'
       '1,2,3\n'
       '4,5,6\n'
       '7,8,9')
```

```python
pd.read_csv(StringIO(int_data),dtype={'col1':int,'col2':float})
```

```python
## Make your own index in the dataframe
## i.e instead of pandas creating the index iteratively. You can use your own index
index_data = ('col1,col2,col3\n'
       '1,2,3\n'
       '4,5,6\n'
       '7,8,9')
## You can tell the column number which you want to use as the index using the index_col parameter
## In this example, the column 0 values will be used as the index
pd.read_csv(StringIO(index_data), index_col = 0)
```
