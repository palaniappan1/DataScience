```python
import matplotlib.pyplot as plt

## Without use of this line, everytime you plot a graph you need to type plt.show
## now with the help if this line you dont need to do that :)
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
## Plot x and y using matplolib
## Scatter plot with x and y (data's) and color (green)
plt.scatter(x,y,c='g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple 2-D Graph')
```

```python
## Just, instead of always getting a straight line, changing values to get something different
y = x*x*x
```

```python
## Plot y versus x as lines and/or markers.
plt.plot(x,y,'r*--')
## plt.plot(x,y) -> Default blue color
## plt.plot(x,y,'r') -> Red color plot
```

```python
## Use of many different parameters
plt.plot(x,y,'r*',linestyle = 'dashed', linewidth = 3, markersize = 15)
```

```python
## Subplots
## plt.subplot(nrows, ncols, pos)
plt.subplot(2,2,1)
plt.plot(x,y,'r') ## I need this diagram in 1st position (3rd argument of the subplot)
plt.subplot(2,2,2)
plt.plot(x,y,'b') ## I need this diagram in 2nd position (3rd argument of the subplot)
plt.subplot(2,2,3)
plt.plot(x,y,'y') ## I need this diagram in 3rd position (3rd argument of the subplot)

## Bcoz we have given 2 rows and 2 columnsm we will get only 4 positions in the grid
```

```python
## Draw a sine curve for fun
## np.arange(start,stop,step)
x = np.arange(0, 4 * np.pi , 0.1)
y = np.sin(x)

plt.title('Sine wave form')
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
## Now with the use of subplot, draw both sine and cosine curves in the same plot
x = np.arange(0, 5*np.pi , 0.1)
y_sine = np.sin(x)
y_cos = np.cos(x)

## Plot the sine curve
plt.subplot(2,2,1)
plt.plot(x,y_sine,'r')
plt.title("Sine Curve")


## Plot the cosine curve
plt.subplot(2,2,2)
plt.plot(x,y_cos,'g')
plt.title("Cosine Curve")
```

```python
## Bar plot
x = [1,2,3]
y = [4,5,6]

x1 = [7,8,9]
y1 = [2,4,6]

plt.bar(x,y)
plt.bar(x1,y1,color = 'r')
```

## Histogram

```python
import collections

x = [10,10,20,10,30,40,50,60,20,40,50]
collections.Counter(x)
```

```python
## With the help of histogram, we can find the frequency( or density or count) 
## of the numbers in the distribution
## Bins is default value of 10, your input values are divided into 10 bins
## [10., 15., 20., 25., 30., 35., 40., 45., 50., 55., 60.] from min 10 to max 60 in your inputs
## It is divided into 10 values
plt.hist(x)
```

```python
## Bar plots
## To learn more about bar plots, see the link below
## https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/box-plot-review#:~:text=A%20box%20and%20whisker%20plot%E2%80%94also%20called%20a%20box%20plot,the%20box%20at%20the%20median.

data = [25,28,29,29,30,34,35,35,37,38]
plt.boxplot(data,vert=True,patch_artist=True)

## The box plot draws box from 1st to 3rd quartile
## Mathematical repr of this example
## Min = 25, Max = 38
## Median = (30 + 34) / 2 = 32
## First Quartile -> Median of datapoints to the left of the median -> 25+28+29+29+30 -> 29
## Third Quartile -> Median of datapoints to the right of the median -> 34+35+35+37+38 -> 35
## The five number summary is 25,29,32,35,38 -> min,1st quartile, median, 3rd quartile, max
```

```python
plt.boxplot(data,vert=False,patch_artist=True)
```

```python
## See the boxplots for multiple datas
new_data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(new_data,vert = True, patch_artist = True)
```

## Pie Chart

```python
labels = ['DSIA','PHS','CA','DDIE']
sizes = [215,245,130,210]
colors = ['gold','green','yellow','red']
explode = [0,.1,0,0] ## How far does the part of the pie chart should be exploded (shown like pie cut)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True, autopct='%1.1f%%')
```
