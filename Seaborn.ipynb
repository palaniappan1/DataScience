```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
## In the above table, tip is dependent on all other features.
```

## Correlation

```python
df.corr()
## This gives all the integer and float values valued columns correlation
## Correlation always vary between -1 and +1
```

```python
sns.heatmap(df.corr())
```

## Distplot

```python
## Dist plot helps us to check the distribution of the columns feature
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
df['tip'].value_counts()
## From the dist plot above, we can find that 2 euros is the most tipped amount.
## We confirmed the same using the value_counts, where 2 euros is paid by most people (33 people)
```

## Bivariate Analysis

```python
## Draw a plot of two variables with bivariate and univariate graphs.
sns.jointplot(x = df['tip'], y = df['total_bill'],kind='hex')
## kind = 'hex' means, the points inside the grid will be in hexagon shape
```

```python
## From this diagram, we can see that people tipped 2 euros max (where the histogram has high peak),
## and the total_bill is between 10 to 20 euros
```

```python
sns.jointplot(x = df['tip'], y = df['total_bill'],kind='reg') ## reg -> regression
```

```python
## Based on the histogam, it draws a PDF (Prbablity Density Function) and also draws a regression 
## line (Best fit line)
```

## Pair Plot

```python
## Plot pairwise relationships in a dataset
## As we saw, there are 3 features (all plotting will be done only for int and float valued features)
## So pair plot is plotted with ine variable in data row with another variable in the row
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = 'sex')
```

## Categorical Plots

```python
## Till now, we saw only plots of int and float values. It's time to see the 
## plot of categorical data points
```

```python
#CountPlot
sns.countplot(x = 'sex',data=df)
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x = 'day',data=df)
```

```python
## Bar Plot
## Show point estimates and errors as rectangular bars
sns.barplot(y = 'total_bill', x = 'sex', data = df)
```

```python
## From this bar plot we can tell that males had more maximum amount of bills than females
```

```python
## Difference between count and bar plot is that in countplot only one feature can be given 
## either in x or y axis (the other axis will have the count value),
## whereas in bar plot two features needs to be given (both x and y axis)
```

```python
## Box Plot
## The plot of 5 number summary
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df)
```

```python
## The dots or rhombus shaped data points are outliers.
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df)
```

```python
## Violin Plot
## It shows distribution both in terms of KDE and Boxplot
```

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
